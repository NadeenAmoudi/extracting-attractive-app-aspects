```python
import pandas as pd
```

```python
from nltk.stem import WordNetLemmatizer
 
lemmatizer = WordNetLemmatizer()
```

```python
from ast import literal_eval
```

```python
data = pd.read_csv('tagged_reviews.csv') 
df = pd.DataFrame(data, columns=['r_id', 's_id', 'app_name','app_id', 'date', 'title', 'review', 'pos', 'rating'])
```

```python
tagged_reviews_list = df['pos'].values.tolist()
```

```python
tagged_reviews_list
```

```python
# Dataframe is converted to list of strings, this code is to remove the strings
for x in range(len(tagged_reviews_list)):
    tagged_reviews_list[x] = literal_eval(tagged_reviews_list[x])
```

```python
# Lemmatization (delivering > deliver)
for x in range(len(tagged_reviews_list)):
    for y in range(len(tagged_reviews_list[x])):
        token, tag = tagged_reviews_list[x][y]
        print("Old token:", token)
        lem_token = lemmatizer.lemmatize(token, tag)
        tagged_reviews_list[x][y] = (lem_token, tag)
        print("New token:", lem_token)
```

```python
tagged_reviews_list
```

```python
table = {"s_id": [], "r_id": [], "app_name": [], "app_id": [], "date": [], "title": [], "review": [], "pos": [], "rating": []}

for x in range(len(df)):
        table["s_id"].append(df['s_id'][x])
        table["r_id"].append(df['r_id'][x])
        table["app_name"].append(df['app_name'][x])
        table["app_id"].append(df['app_id'][x])
        table["date"].append(df['date'][x])
        table["title"].append(df['title'][x])
        table["review"].append(df['review'][x])
        table["pos"].append(tagged_reviews_list[x])
        table["rating"].append(df['rating'][x])
```

```python
df=pd.DataFrame.from_dict(table,orient='index').transpose()
df.to_csv('tagged_lemmatized_reviews.csv', index = False)
```
