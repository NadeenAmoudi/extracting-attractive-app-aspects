# Preprocessing 2 

```python
data = pd.read_csv('splitted_reviews_into_sentences_sID.csv') 
df = pd.DataFrame(data, columns=['r_id', 's_id','app_name','app_id', 'date', 'title', 'review', 'rating'])
```

```python
# Convert the data frame to list of strings
reviews_list = df['review'].astype(str).values.tolist()
```

```python
for x in range(len(df)):
        reviews_list[x] = nfx.fix_contractions(reviews_list[x])
        reviews_list[x] = nfx.remove_puncts(reviews_list[x])
#         reviews_list[x] = nfx.remove_stopwords(reviews_list[x])
# Remove special characters has removed the Arabic text
        reviews_list[x] = nfx.remove_special_characters(reviews_list[x]) 
        reviews_list[x] = nfx.remove_emojis(reviews_list[x])
```

```python
reviews_list
```

```python
# Convert the lists to dataframe
dfReview = pd.DataFrame (reviews_list, columns = ['review'])
```

```python
table = {"r_id": [], "s_id": [], "app_name": [], "app_id": [], "date": [], "title": [], "review": [], "rating": []}

for x in range(len(df)):
    table["r_id"].append(df['r_id'][x])
    table["s_id"].append(df['s_id'][x])
    table["app_name"].append(df['app_name'][x])
    table["app_id"].append(df['app_id'][x])
    table["date"].append(df['date'][x])
    table["title"].append(df['title'][x])
    table["review"].append(dfReview['review'][x])
    table["rating"].append(df['rating'][x])
```

```python
df=pd.DataFrame.from_dict(table,orient='index').transpose()
df.to_csv('preprocessing_2.csv', index = False)
```
