```python
data = pd.read_csv('features_new_weight.csv') 
df = pd.DataFrame(data, columns=['feature',  'sentence_sentiment'])
```

```python
feature_to_match = df['feature'].to_list()
feature_sentiment = df['sentence_sentiment'].to_list()
```

```python
features_list
```

```python
satisfaction = []
dissatisfaction = []
neutral = []
freq = []
for x in range(len(features_list)):
    frequency = n_frequency = n_total_score = p_frequency = p_total_score = np_frequency = np_total_score = 0
    for y in range(len(feature_to_match)):
        if features_list[x] == feature_to_match[y]:
            frequency = frequency + 1
            if feature_sentiment[y] < 0:
                n_frequency = n_frequency + 1
                n_total_score = n_total_score + abs(feature_sentiment[y])
            elif feature_sentiment[y] > 0:
                p_frequency = p_frequency + 1
                p_total_score = p_total_score + feature_sentiment[y]
            elif feature_sentiment[y] == 0:
                np_frequency = np_frequency + 1
                np_total_score = np_total_score + 1
    freq.insert(x, frequency)
    
    if n_frequency > 0:
        n_avg = float(n_total_score*weight[x])/frequency
        dissatisfaction.insert(x, n_avg)
    elif n_frequency == 0:
        dissatisfaction.insert(x, 0)
        
    if p_frequency > 0:
        p_avg = float(p_total_score*weight[x])/frequency
        satisfaction.insert(x, p_avg)
    elif p_frequency == 0:
        satisfaction.insert(x, 0)
    
    if np_frequency > 0:
#         print(np_frequency)
        np_avg = float(np_total_score*weight[x])/frequency
        neutral.insert(x, np_avg)
    elif np_frequency == 0:
        neutral.insert(x, 0)
```

```python
table = {"feature": [], "dissatisfaction": [], "satisfaction": [], "neutral": [], "frequency": []}

for x in range(len(features_list)):
    table["feature"].append(features_list[x])
    table["dissatisfaction"].append(dissatisfaction[x])
    table["satisfaction"].append(satisfaction[x])
    table["neutral"].append(neutral[x])
    table["frequency"].append(freq[x])
```

```python
df=pd.DataFrame.from_dict(table,orient='index').transpose()
df.to_csv('features_dissatisfaction_satisfaction_neutral_NEW.csv', index = False)
```
